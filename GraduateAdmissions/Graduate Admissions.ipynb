```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Changing the size of the graphs
sns.set(rc={'figure.figsize':(16, 8)})
```

```python
#Admissions Data Set
admissions = pd.read_csv(r"C:\Users\sarat/Videos/DSS/Admission_Predict_Ver1.1.csv")
admissions
```

```python
admissions.columns
```

```python
#Are GRE and TOEFL scores positively correlated?
sns.regplot(x='GRE Score', y='TOEFL Score', data=admissions, x_jitter = 2, y_jitter = 2);
plt.xlabel('GRE Score')
plt.ylabel('TOEFL Score')
plt.title('GRE Score vs TOEFL Score');
```

```python
admissions.columns
```

```python
#Do higher GRE, TOEFL or test scores increase the chance of getting into a Masters program
sns.regplot(x='GRE Score', y='Chance of Admit ', data=admissions, x_jitter = 3.4, y_jitter = 0.01, scatter_kws={"color":"blue","alpha":0.3,"s":40});
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit ')
plt.title('GRE Score vs Chance of Admit');
```

```python
#Do higher GRE, TOEFL or test scores increase the chance of getting into a Masters program
sns.regplot(x='TOEFL Score', y='Chance of Admit ', data=admissions, x_jitter = 1.2, y_jitter = 0.01, scatter_kws={"color":"blue","alpha":0.3,"s":40});
plt.xlabel('TOEFL Score')
plt.ylabel('Chance of Admit ')
plt.title('TOEFL Score vs Chance of Admit');
```
