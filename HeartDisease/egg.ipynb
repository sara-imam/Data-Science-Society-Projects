```python
def distance(pt1, pt2):
    """Return the distance between two points, represented as arrays"""
    return np.sqrt(sum((pt1 - pt2)**2))
```

```python
def row_distance(row1, row2):
#     print(len(row1), len(row2))
    """Return the distance between two numerical rows of a table"""
    return distance(np.array(row1), np.array(row2))
```

```python
def distances(training, unknown_person):
    """Compute distance between our unknown person and every row in training.
    Return training augmented with Distance column"""
    distances = []
    training_table_copy = training.copy(deep = True)
    attributes_only = training_table_copy.drop(columns = ["target"])
#     print(attributes_only.columns)
    
    for i in np.arange(0, attributes_only.shape[0]):
        something = attributes_only.iloc[i]
        distances += [row_distance(something, unknown_person)]
        
    training_table_copy['Distance_to_unknown'] = distances
        
    return training_table_copy
```

```python
def closest(training, unknown_person, k):
    """Return a table of the k closest neighbors to example"""
    return distances(training, unknown_person).sort_values('Distance_to_unknown').iloc[0:k]
```

```python
def majority_class(topk):
    """Return the class with the highest count"""
    return topk.groupby("target")["target"].count().sort_values(ascending=False).index[0]
```

```python
def classify(training, unknown_person, k):
    "Return the majority class among the k nearest neighbors of example"
    return majority_class(closest(training, unknown_person, k))
```

```python
def evaluate_accuracy(training, test, k):
    """Return the proportion of correctly classified examples 
    in the test set"""
    test_attributes = test.drop(columns = ["target"])
    num_correct = 0
    for i in np.arange(test.shape[0]):
        c = classify(training, test_attributes.iloc[i], k)
        if (c == test["target"].iloc[i]):
            num_correct +=  1
    return num_correct / len(test.index)
```
